#### Construct a function to return the companies we would like to know with chosen category and its value.

```python
# Return a data frame containing the top n rows of companies, 
# based on revenues for a specific value in the category,only if both category and value are provided. 
# If both category and value are not provided, the top n companies overall should be displayed.

def category_value_func(df,category='None',value='None',n=10):
    
    # if category and value both have values
    if (category!='None') and (value!='None'):
          
        # subset the group by category and value
        df_category_str = df[df[category]==value]
            
        # sort data by revenue
        df_category_str_sort = df_category_str.sort_values(['REVENUE'],ascending=False)
            
        # return first n rows
        return df_category_str_sort.iloc[:n]
        
    else:
            # sort data by revenue
            df = df.sort_values(['REVENUE'],ascending=False)
            
            # return first n rows
            return df.iloc[:n]
```

#### 25% percentile and 50% percentile with their awards
Any company who is in an industry with a "Services" name included as part of its industry. <br>
Top quarter of these companies will be given a Trailblazer award<br>
The top half will be given a Pioneer award.<br>
Showing the company, its rank, its revenue, and which award it will earn, sorted by revenue in descending order.<br>

```python
# get data with servies related industry
services_company = inc5000[inc5000['INDUSTRY'].str.contains("Services")]
```

```python
# get 75% percentile revenue based on their revenue
percentile_75 = np.percentile(services_company['REVENUE'],75)
```

```python
# get 50% percentile revenue based on their revenue
percentile_50 = np.percentile(services_company['REVENUE'],50)
```

```python
# get top quater of companies based on their revenue
top25_percentile = services_company[services_company['REVENUE']>=percentile_75]
```

#### helpful function to dertermine what awards should go to those award-wining companies



```python
# revenue of servieces type companies
service_revenue = services_company['REVENUE']

def awards(services_company):
    
    service_revenue = services_company['REVENUE']
    
    # company with Trailblazer award
    if service_revenue >= percentile_75:
        return 'Trailblazer'
    
    # company with Pioneer award
    elif (service_revenue < percentile_75) and (service_revenue <= percentile_50):
        return 'Pioneer'
    
    # company don't win any award
    else:
        return 
    
# make a copy to avoid SettingWithCopyWarning warning
services_company = services_company.copy()

# add new column with the value of awards specified from the upper function
services_company['AWARDS'] = services_company.apply(awards, axis=1)
```

```python
# name the index 
services_company.index.name = 'Rank'

# get only the useful columns and sort by revenue in descending order
services_company_award = services_company[['COMPANY NAME','REVENUE','AWARDS']].sort_values(
    ['REVENUE'],ascending=False)
services_company_award
```
